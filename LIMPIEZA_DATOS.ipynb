##### Se crea una función que nos permita extraer las variables en común de cada año.

```python
columnas_df = ['ANO_EMISION', 'MES_EMISION', 'DIA_EMISION', 'DEPTO_MCPIO', 'QUIEN_REPORTA', 'VIC_SEXO', 'VIC_EDAD', 'TOTAL_HIJOS', 'NUM_HIJ_HOM', 'NUM_HIJ_MUJ', 'VIC_ALFAB',
            'VIC_ESCOLARIDAD', 'VIC_EST_CIV', 'VIC_GRUPET', 'VIC_NACIONAL', 'VIC_TRABAJA', 'VIC_OCUP', 'VIC_DEDICA', 'VIC_DISC', 'TIPO_DISCAQ', 'VIC_REL_AGR',
            'OTRAS_VICTIMAS', 'VIC_OTRAS_HOM', 'VIC_OTRAS_MUJ', 'VIC_OTRAS_N_OS', 'VIC_OTRAS_N_AS', 'HEC_DIA', 'HEC_MES', 'HEC_ANO', 'HEC_DEPTOMCPIO', 'HEC_AREA',
            'HEC_TIPAGRE', 'HEC_RECUR_DENUN', 'INST_DONDE_DENUNCIO', 'AGR_SEXO', 'AGR_EDAD', 'AGR_ALFAB', 'AGR_ESCOLARIDAD', 'AGR_EST_CIV', 'AGR_GRUPET', 'AGR_NACIONAL',
            'AGR_TRABAJA', 'AGR_OCUP', 'AGR_DEDICA', 'INST_DENUN_HECHO', 'MEDIDAS_SEGURIDAD']
def preparar_anio_datos(ruta, columnas = columnas_df):

    cambios_nombres = {'AGR_GURPET': 'AGR_GRUPET'}
    
    df = pd.read_spss(ruta)
    df = df.rename(columns = cambios_nombres)
    
    df = df[columnas]

    return df
```

##### Se realiza la concatenación de las bases de datos de cada año.

```python
anios = ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
df = preparar_anio_datos('DATOS_PROYECTO_1/' + anios[0] + '.sav')

for i in range(1, len(anios)):
    df = pd.concat([df, preparar_anio_datos('DATOS_PROYECTO_1/' + anios[i] + '.sav')])

df.tail(2)
```
